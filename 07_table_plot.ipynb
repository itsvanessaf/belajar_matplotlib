```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

Sample Data

```python
data_uas=[
    ['Vanessa', 70],
    ['Rien', 83],
    ['Chelsea', 62],
    ['Devi', 74],
    ['Alia', 71]
]
```

```python
data_uas
```

Table Plot dengan OO Style

```python
fig, ax = plt.subplots()

table = plt.table(cellText=data_uas, 
                  loc='center')

table.set_fontsize(14)
table.scale(1, 4) 

ax.axis(False)

plt.show()
```

Table Plot dengan pyplot Style

```python
table = plt.table(cellText=data_uas, 
                  loc='center')

table.set_fontsize(14)
table.scale(1, 4) 

ax = plt.gca()
ax.axis(False)

plt.show()
```
